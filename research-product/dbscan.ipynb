## Sample implementation of DBScan

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

### Example using random blobs (2-Dimensional data)

```python
centers = [[1, 0.5], [2, 2], [1, -1]]
stds = [0.1, 0.4, 0.3]
X, labels_true = make_blobs(n_samples=1000, centers=centers, cluster_std=stds, random_state=42)
fig = plt.figure(figsize=(10, 10))
sns.scatterplot(data = X, x=X[:,0], y=X[:,1])
plt.show()
```

```python
db = DBSCAN(eps=.25, min_samples=5).fit(X) 
labels = db.labels_
fig = plt.figure(figsize=(10, 10))
sns.scatterplot(X[:,0], X[:,1], hue=["cluster-{}".format(x) for x in labels])
plt.show()
```

### Example using multi-dimensional data

```python
random_data = np.random.randint(0,100,size=(1000,6))
df = pd.DataFrame(random_data, columns=['Column 1','Column 2','Column 3', 'Column 4','Column 5','Column 6'])
```

```python
random_data = np.random.randint(0,100,size=(1000,3))
df = pd.DataFrame(random_data, columns=['Column 1','Column 2','Column 3'])

# transforming df into 2D data
pca = PCA(2)
df = pca.fit_transform(df)

db = DBSCAN(eps=5, min_samples=4).fit(df) 
label = db.labels_
u_labels = np.unique(label)
fig = plt.figure(figsize=(10, 10))
for i in u_labels:
        plt.scatter(df[label == i,0], df[label == i, 1], label = i)
plt.legend()
plt.show()
```
